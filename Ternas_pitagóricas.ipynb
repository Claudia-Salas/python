Tomamos el cuadrado de cada elemento uilitizando comprensiones

```python
x=[i**2 for i in range(1001)]
y=[i**2 for i in range(1001)]
z=[i**2 for i in range(1001)]

def terna(a,b,c):                          #definimos a la terna cualquiera como (a,b,c)
  terna_pitagórica=[]
  if a+b==c:
    terna_pitagórica.append((a,b,c))
    print(a,b,c)

for j in range(1001):
    for k in range (1001):
      for m in range (1001):
        terna(x[j],y[k],z[m])
```
